```python
# ==========================================
# üöÄ MASTER SCRIPT: PREP + TRAIN + TEST (FIXED)
# ==========================================
import os
import sys

# --- PART 1: SETUP & INSTALL ---
print("üõ†Ô∏è Installing libraries...")
!pip install mne==1.6.0 scipy matplotlib h5py pandas

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import mne
from scipy import signal
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from PIL import Image

# Global Settings
WORK_DIR = "/content"
EDF_FOLDER = os.path.join(WORK_DIR, "raw_data")
OUTPUT_FOLDER = os.path.join(WORK_DIR, "dataset_spectrograms")
WINDOW_SECONDS = 4
FS = 256

# --- PART 2: DOWNLOAD DATA ---
print("\n‚¨áÔ∏è DOWNLOADING DATA...")
os.makedirs(EDF_FOLDER, exist_ok=True)
os.makedirs(os.path.join(OUTPUT_FOLDER, "healthy"), exist_ok=True)
os.makedirs(os.path.join(OUTPUT_FOLDER, "seizure"), exist_ok=True)

files_to_download = [
    "chb01/chb01_03.edf", "chb01/chb01_04.edf",
    "chb02/chb02_16.edf", "chb03/chb03_01.edf"
]
base_url = "https://physionet.org/files/chbmit/1.0.0/"

for f in files_to_download:
    local_path = os.path.join(EDF_FOLDER, os.path.basename(f))
    if not os.path.exists(local_path):
        os.system(f"wget -q -O {local_path} {base_url}{f}")

# --- PART 3: GENERATE SPECTROGRAMS ---
print("\nüöÄ GENERATING SPECTROGRAMS (This ensures the folder exists!)...")
SEIZURE_TIMES = {
    "chb01_03.edf": [(2996, 3036)], "chb01_04.edf": [(1467, 1494)],
    "chb02_16.edf": [(130, 212), (2972, 3053)], "chb03_01.edf": [(362, 414)]
}

def save_spectrogram(data, label, filename, index):
    try:
        f, t, Sxx = signal.spectrogram(data, fs=FS, nperseg=128, noverlap=64)
        fig = plt.figure(figsize=(1.28, 1.28), dpi=100)
        plt.pcolormesh(t, f, 10 * np.log10(Sxx + 1e-10), shading='gouraud', cmap='jet')
        plt.axis('off')

        class_folder = "seizure" if label == 1 else "healthy"
        save_path = os.path.join(OUTPUT_FOLDER, class_folder, f"{filename}_{index}.png")
        plt.savefig(save_path, bbox_inches='tight', pad_inches=0)
        plt.close(fig)
    except:
        plt.close(fig)

files = [f for f in os.listdir(EDF_FOLDER) if f.endswith('.edf')]
for file_name in files:
    print(f"  Processing {file_name}...")
    try:
        raw = mne.io.read_raw_edf(os.path.join(EDF_FOLDER, file_name), preload=True, verbose=False)
        raw.notch_filter(freqs=60, method='fir', verbose=False)
        raw.filter(1., 70., verbose=False)
        data = raw.get_data()[0] * 1e6

        n_windows = len(data) // (WINDOW_SECONDS * FS)
        file_seizures = SEIZURE_TIMES.get(file_name, [])

        for i in range(n_windows):
            start = i * WINDOW_SECONDS * FS
            end = start + (WINDOW_SECONDS * FS)
            mid_sec = (start + end) / 2 / FS

            is_seizure = 0
            for (s_start, s_end) in file_seizures:
                if s_start <= mid_sec <= s_end:
                    is_seizure = 1; break

            if is_seizure == 1 or (i % 10 == 0):
                save_spectrogram(data[start:end], is_seizure, file_name.replace('.edf',''), i)
    except: continue

# --- PART 4: TRAIN MODEL ---
print("\nüß† TRAINING CNN MODEL...")
# Check if data exists before crashing
if len(os.listdir(os.path.join(OUTPUT_FOLDER, "healthy"))) == 0:
    raise RuntimeError("‚ùå CRITICAL: No images were generated. Check download step.")

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)
train_gen = datagen.flow_from_directory(
    OUTPUT_FOLDER, target_size=(128, 128), batch_size=32,
    class_mode='binary', color_mode='grayscale', subset='training', shuffle=True
)
val_gen = datagen.flow_from_directory(
    OUTPUT_FOLDER, target_size=(128, 128), batch_size=32,
    class_mode='binary', color_mode='grayscale', subset='validation', shuffle=False
)

model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(128, 128, 1)),
    MaxPooling2D(2, 2),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D(2, 2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(train_gen, validation_data=val_gen, epochs=15)

# --- PART 5: VISUALIZE RESULTS ---
print("\nüìà GENERATING REPORTS...")
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
plt.figure(figsize=(8, 4))
plt.plot(acc, label='Training Acc'); plt.plot(val_acc, label='Validation Acc')
plt.title('Accuracy'); plt.legend(); plt.show()

# --- PART 6: BLIND TEST ---
print("\nüè• RUNNING BLIND TEST (chb10_12.edf)...")
test_path = "chb10_12.edf"
if not os.path.exists(test_path):
    os.system("wget -q https://physionet.org/files/chbmit/1.0.0/chb10/chb10_12.edf")

raw = mne.io.read_raw_edf(test_path, preload=True, verbose=False)
raw.filter(1., 70., verbose=False)
data = raw.get_data()[0] * 1e6

predictions, ground_truth = [], []
REAL_SEIZURE_START, REAL_SEIZURE_END = 6313, 6348
n_windows = len(data) // (WINDOW_SECONDS * FS)

print(f"üîé Scanning {n_windows} windows...")
for i in range(n_windows):
    start = i * WINDOW_SECONDS * FS
    end = start + (WINDOW_SECONDS * FS)
    segment = data[start:end]

    f, t, Sxx = signal.spectrogram(segment, fs=FS, nperseg=128, noverlap=64)
    Sxx_log = 10 * np.log10(Sxx + 1e-10)
    img_norm = (Sxx_log - Sxx_log.min()) / (Sxx_log.max() - Sxx_log.min())
    img_pil = Image.fromarray(np.uint8(img_norm * 255)).resize((128, 128))
    img_input = np.array(img_pil).reshape(1, 128, 128, 1) / 255.0

    pred = model.predict(img_input, verbose=0)
    predictions.append(1 if pred[0][0] > 0.7 else 0)

    mid_sec = (start + end) / 2 / FS
    ground_truth.append(1 if REAL_SEIZURE_START <= mid_sec <= REAL_SEIZURE_END else 0)

print(f"\n‚úÖ FINAL TEST ACCURACY: {accuracy_score(ground_truth, predictions)*100:.2f}%")
print(classification_report(ground_truth, predictions, target_names=['Healthy', 'Seizure']))
```
